```python
def get_col_row(df:pd.DataFrame):
    row = 0 
    col = 0
    for value in df.itertuples():
        row,col1,col2,col3 = value
        if str(col1).strip().upper()=="ACTIVO":
            col = 0
            break
        elif str(col2).strip().upper()=="ACTIVO":
            col = 1
            break
        elif str(col3).strip().upper()=="ACTIVO":
            col = 2
            break
    return col,row
```

```python
def get_data_total(df):
    column_total = [i for i in range(len(df.columns)) if str(df.iloc[1, i]).strip().upper() == 'TOTAL']
    nombres = df.iloc[0, [i - 2 for i in column_total]].reset_index(drop=True)
    df.iloc[0, column_total] = nombres
    df.iloc[0, [i - 2 for i in column_total]] = np.nan
    df = df.dropna(axis=1, subset=[0], how='any').reset_index(drop=True)
    df.iloc[1,0] = "MONEDA"
    df.columns = df.iloc[0]
    df.drop(0,inplace=True)
    df = df.loc[:, ~df.columns.duplicated()]
    return df
```

```python
# df = df.iloc[row:,col:].reset_index(drop=True).copy(deep=True)
# df = get_data_total(df)

def get_rows(df:pd.DataFrame):

    rows = [value for value in df["ACTIVO"] 
            if pd.notna(value) 
            and "BALANCE" not in str(value) 
            and "TIPO DE" not in str(value) 
            and "EN MILES DE" not in str(value) 
            and "A PART" not in str(value) 
            and "CON RELACION" not in str(value) 
            and value not in reporte_mes1
        #     and value not in reporte_mes2
            and not isinstance(value, datetime)  
            and not es_fecha(value)
            ]  
    return rows

```

```python
import pandas as pd

def get_data_complete(filepath:str,entity,year):
    # Leer el archivo Excel
    if valid_month_year(filepath):
        return
    df = pd.read_excel(filepath, header=None)
    
   
    df = upper_data(df)
    df.dropna(axis=1, how="all", inplace=True)
    df.dropna(axis=0, how="all", inplace=True)
    df = df.reset_index(drop=True).copy(deep=True)

    col, row = get_col_row(df.iloc[:20, :3])
    df = df.iloc[row:, col:].reset_index(drop=True).copy(deep=True)

    df = get_data_total(df)
    rows = get_rows(df)
    df1 = df[df["ACTIVO"].isin(rows)].reset_index(drop=True).copy(deep=True)
    filename = filepath.split("/")[-1].split(".")[0]
    directory = f"data_clean/{entity}/BG/{year}/"
    os.makedirs(directory,exist_ok=True)
    new_filename = f"{directory}/{filename}"
    df1.to_excel(f"{new_filename}.xlsx", index=False)

```

```python
for entity in vars.ENTITYS:
    for year in vars.YEARS:
        fileptahs = f'datasets/{entity}/EF/{year}'
        for file in os.listdir(fileptahs):
            filepath = f"{fileptahs}/{file}"

            get_data_complete(filepath,entity,year)
        print(f"Archivos del {year} processados")
```
